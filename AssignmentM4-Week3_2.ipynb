```python
# import libraries

import pandas as pd

# define url and load the data /webscraping the data

url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df=pd.read_html(url)

df[0].head()
df1=df[0]
df1
```

```python
# df2 = Dataframe filtered by Boroughs that are not assinged

df2 = df1['Borough'] != 'Not assigned'
#print(df2)

df2=df1[df2]
# reset index
df2=df2.reset_index(drop=True)
df2

# replace missing Neighbourhoods by the name of the Borough

df2['Neighbourhood'].replace(['Not assigned'], [df2['Borough']], inplace=True)
df3= df2.groupby(['Postcode','Borough'])['Neighbourhood'].transform(lambda x:', '.join(x))
df4= df2.merge(df3, left_index= True, right_index = True)
df5= df4[['Postcode', 'Borough', 'Neighbourhood_y']]
df5= df5.drop_duplicates()
df5
```

```python
df_geo = pd.read_csv(r'C:\Users\TNguyen\Downloads\Geospatial_Coordinates.csv')
df_geo

df_geo.rename(columns={'Postal Code':'Postcode'}, inplace=True)
df_geo
```

```python
df6= pd.merge(df5, df_geo, on='Postcode', how ='left')
df6.rename(columns={'Neighbourhood_y':'Neighbourhood'}, inplace=True)
df6
```
