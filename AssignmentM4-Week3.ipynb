```python
# import libraries

import pandas as pd
```

```python
# define url and load the data /webscraping the data

url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df=pd.read_html(url)
```

```python
df[0].head()
df1=df[0]
```

```python
df1
```

```python
# df2 = Dataframe filtered by Boroughs that are not assinged

df2 = df1['Borough'] != 'Not assigned'
#print(df2)

df2=df1[df2]
df2
```

```python
# reset index
df2=df2.reset_index(drop=True)
df2
```

```python
# replace missing Neighbourhoods by the name of the Borough

df2['Neighbourhood'].replace(['Not assigned'], [df2['Borough']], inplace=True)
df2
```

```python
# group by postcode and transform the column Neighbourhood

df3= df2.groupby(['Postcode','Borough'])['Neighbourhood'].transform(lambda x:', '.join(x))
df3
```

```python
# create new dataframe 
df4= df2.merge(df3, left_index= True, right_index = True)
df4
```
